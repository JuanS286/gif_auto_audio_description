```python
import pandas as pd
from collections import Counter
```

```python
# Load the dataset
df = pd.read_csv('/content/drive/My Drive/tgif-v1.0.tsv', sep='\t', header=None, names=['gif_url', 'description'])
```

```python
# Define categories and associated keywords
categories = {
    'Animals': ['cat', 'dog', 'animal', 'mouse', 'fish', 'cow', 'elephant', 'kitten', 'puppy', 'pup', 'rat','wolf','hamster','lion','bear','gorilla','dinosaur',
                'squirrel','bird','tiger','rabbit','chipmunk','duck','fox','panda','pug','lizard','golden retriever','pigeon','turtle','sloth','horse',
                'ferret','snake','monkey','shark','penguin','goat','dragon','whale'],
    'Humans': ['man', 'woman', 'boy', 'girl', 'person', 'men', 'women', 'child', 'reporter', 'children', 'people', 'male', 'female', 'santa claus',
                'crowd', 'dancer', 'teenager', 'policeman', 'skateboarder', 'officer', 'officers', 'kid','husband', 'wife','guy','lady',
                'someone','passenger','biker','motorcyclist','cyclist','baby','singer','couple','actress','musician','lovers','player','guitarist',
                'soldiers','ballerina','boxer','somebody','model','captain','rednecks','competitors','performer','ninjas','transsexual','dictator',
                'asian','redhead','vampire','samurai','drummer','friend','rocker','avengers','footballers','band','actor','she','he', 'they','wrestler',
                'gymnast','superhero','bunny','cheerleaders','students','snowboarder','fighter','rock group','mermaids','ladies','drag queen','pilot','teacher'],
    'Objects': ['car', 'skateboard', 'bike', 'vehicle', 'object', 'moon', 'bus','skate','glove','glass','shelf','cigarette','punch','tap','pint','beer',
                'swords','silhouettes','device','alien','machines','television','motorcycles','creature','helicopter','robot'],
    'Text': ['text', 'words', 'message', 'say', 'shout'],
    'Animations': ['animation', 'cartoon'],
    'Famous':['paul', 'arnold','beatles','jhon','lennon','mccartney','Michael','Jackson','lucille','kobe','bryant'],
    'Insects':['butterfly','bee','monarch butterfly','spider']
}
```

```python
# Function to categorize each description
def categorize_description(description):
    for category, keywords in categories.items():
        if any(keyword in description.lower() for keyword in keywords):
            return category
    return 'Uncategorized'
```

```python
# Apply the function to the dataset
df['category'] = df['description'].apply(categorize_description)
```
