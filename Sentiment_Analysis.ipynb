**Sentiment Analysis**

```python
from textblob import TextBlob

# Define a function for sentiment analysis
def get_sentiment(text):
    blob = TextBlob(text)
    # Polarity ranges from -1 (negative) to 1 (positive)
    return blob.sentiment.polarity

# Apply the sentiment analysis function to the GIF descriptions
df['Sentiment'] = df['Description'].apply(get_sentiment)

def get_sentiment_category(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity

    if polarity > 0.05:
        return "Positive"
    elif polarity < -0.05:
        return "Negative"
    else:
        return "Neutral"

# Apply the sentiment analysis function to the GIF descriptions
df['Sentiment_Category'] = df['Description'].apply(get_sentiment_category)

# Check the result
df[['Description', 'Sentiment', 'Sentiment_Category']].head()
```

Polarity: Ranges from -1 (negative sentiment) to 1 (positive sentiment).
For example, a description like "a man is smiling" might yield a positive sentiment, while "a cat is scared" might yield a negative sentiment.

A polarity of 0.0 means that the text is considered neutral. It does not express strong positive or negative emotions. For example, a neutral sentence might state a fact or provide information without showing any sentiment, such as:

"The sky is blue."
"The meeting is scheduled for tomorrow."

```python
import nltk

# Function to extract the first two present or past tense verbs excluding 'is' and 'are'
def extract_verbs(description):
    words = nltk.word_tokenize(description)  # Tokenize the sentence
    pos_tags = nltk.pos_tag(words)  # POS tagging

    verbs = []
    for i, (word, pos) in enumerate(pos_tags):
        # Check for present tense (VB, VBP, VBZ), past tense (VBD), or present continuous (VBG)
        if (pos in ['VB', 'VBP', 'VBZ', 'VBD'] and word.lower() not in ['is', 'are']):
            verbs.append(word)
        elif pos == 'VBG':  # Identify gerund forms
            # Check for the auxiliary verb (am, is) in the previous tokens
            if i > 0 and pos_tags[i - 1][0].lower() in ['am', 'is']:
                verbs.append(word)
            # Allow gerunds that may not be immediately preceded by an auxiliary
            elif len(verbs) < 2:  # Allow adding if we don't have two actions yet
                verbs.append(word)

        # Include specific verbs like standing, turning, jumping, etc.
        if word.lower() in ['standing', 'turning', 'jumping'] and word.lower() not in ['is', 'are']:
            verbs.append(word)

        # Stop after extracting the first two verbs
        if len(verbs) == 2:
            break

    return verbs

# Apply the verb extraction function to the Description column
df['Actions'] = df['Description'].apply(extract_verbs)

# Create the Action_1 and Action_2 columns
df['Action_1'] = df['Actions'].apply(lambda x: x[0] if len(x) > 0 else None)
df['Action_2'] = df['Actions'].apply(lambda x: x[1] if len(x) > 1 else None)

# Keep only the necessary columns: URL, Description, Category, Action_1, Action_2
df = df[['URL', 'Description', 'Category', 'Sentiment', 'Sentiment_Category', 'Action_1', 'Action_2']]

# Display the resulting DataFrame
print(df)
```

The NLTK (Natural Language Toolkit) library is a powerful suite of tools for working with human language data (text). It provides libraries for various tasks like tokenization, part-of-speech (POS) tagging, and more.
